```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
import pandas as pd
```

```python
pkl = "250419_s3w0iwik_eval/epoch0/medrmsd_epoch0_loss.pkl"
with open(pkl, 'rb') as file:
    med_data = pickle.load(file)

pkl = "250419_s3w0iwik_eval/epoch0/minrmsd_epoch0_loss.pkl"
with open(pkl, 'rb') as file:
    min_data = pickle.load(file)
```

```python
# Extract the bins and their corresponding data
categories = ['tr', 'rot', 'tor']  # Categories to plot
ylim = [(0,7), (0,4), (0,7)]
rows = np.arange(0, 1.0, 0.05)  # Rows: 0, 0.05, 0.1, ..., 1.0

# Create a figure with subplots
fig, axes = plt.subplots(len(categories), 1, figsize=(6, 10), sharex=True)

for i, (category, y) in enumerate(zip(categories, ylim)):
    bins = list(min_data[category].keys())
    
    # Initialize a dictionary to store the averages
    averages = {bin_name: [] for bin_name in bins}

    # Compute the averages for each bin
    for bin_name in bins:
        for row in rows:
            # Assuming the data in each bin is a list of lists or similar structure
            # Replace `data[category][bin_name]` with the actual structure if different
            min_data[category][bin_name][row] = [val for val in min_data[category][bin_name].get(row, []) if val <= 100]
            averages[bin_name].append(np.mean(min_data[category][bin_name].get(row, [])))

    # Create a DataFrame for the table
    df_averages = pd.DataFrame(averages, index=rows)

    # Plot the data for the current category on its subplot
    ax = axes[i]
    for column in df_averages.columns:
        df_averages[column].plot(ax=ax, label=column)
    ax.set_ylim(y)

    # Add labels, title, and legend
    ax.set_xlabel("Denoising steps", fontsize=12)
    ax.set_ylabel(f"{category.capitalize()} loss", fontsize=12)
    ax.set_title(f"{category.capitalize()} Loss vs Noise")
    # break
axes[1].legend(title="RMSDs", loc='upper left', bbox_to_anchor=(1, 1))
```
